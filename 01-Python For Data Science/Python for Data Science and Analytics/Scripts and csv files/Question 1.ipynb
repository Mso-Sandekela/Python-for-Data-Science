```python
#==================================================
#******************Question 1********************
#--------------------------------------------------
#Student Name: Sopumelela
#Student Surname: Sandekela
#Student Number: CON-1475940-C5L6
#==================================================

#importing libraries.
import numpy as np
import pandas as pd
import random

np.random.seed(0)

#==============================================================================
#THIS FUNCTION CREATE A WHOLE DATASET...
#==============================================================================
def create_DataFrame():
    #------------------------------------------------------------------
    #Generating student number...
    #------------------------------------------------------------------
    def generate_Stud_Num():
        import string

        studentNum = ""
        alphabets = string.ascii_uppercase
        digits = string.digits

        #---------------------------------------------
        #Function for generating random string
        #---------------------------------------------
        def generate_alphabets(line,num):
            line += "".join([random.choice(alphabets) 
                             for i in range(num)])

            return line
        #---------------------------------------------

        #---------------------------------------------
        #Function for generating random Integers
        #---------------------------------------------
        def generate_Integers(line):
            line += "".join([random.choice(digits) 
                             for i in range(3)])

            return line
        #---------------------------------------------

        #CON537GF
        studStr = generate_alphabets(studentNum, 3)#CON

        studInt = generate_Integers(studStr)#537

        studentNum = generate_alphabets(studInt, 2)#GF

        #Adding "-" to Student number.
        #CON-537-GF
        stud = studentNum
        studentNum = stud[0]+stud[1]+stud[2] 
        studentNum += '-' + stud[3]+stud[4]+stud[5]
        studentNum += '-' +stud[6]+stud[7]

        return studentNum
        #------------------------------------------------------------------

    x=0
    studNum = np.array([])
    while x != 150:
        studentNumber = ""
        studentNumber = generate_Stud_Num()
        studNum = np.append(studNum, studentNumber)
        x += 1


    #------------------------------------------------------------------
    #Generating student ages using random()...
    #------------------------------------------------------------------
    #my goal is to create a function that generate numpy array 
    #with random int elements everytime i call it.

    def generate_ages():
        age_18to25 = np.array([])
        age_25to35 = np.array([])
        age_35to45 = np.array([])
        age_over45 = np.array([])
        
        for j in range(150):
            rand = random.randint(1, 4)
            if (rand == 1):
                age = np.random.randint(18, 26) #18-25
                age_18to25 = np.append(age_18to25, age)
                age_25to35 = np.append(age_25to35, 0)
                age_35to45 = np.append(age_35to45, 0)
                age_over45 = np.append(age_over45, 0)
                
            elif (rand == 2):
                age = np.random.randint(24, 36) #25-35
                age_25to35 = np.append(age_25to35, age)
                age_18to25 = np.append(age_18to25, 0)
                age_35to45 = np.append(age_35to45, 0)
                age_over45 = np.append(age_over45, 0)
            
            elif (rand == 3):
                age = random.randint(36, 46) #35-45
                age_35to45 = np.append(age_35to45, age)
                age_18to25 = np.append(age_18to25, 0)
                age_25to35 = np.append(age_25to35, 0)
                age_over45 = np.append(age_over45, 0)
        
            elif (rand == 4):
                age = random.randint(45, 60) #35-45
                age_over45 = np.append(age_over45, age)
                age_18to25 = np.append(age_18to25, 0)
                age_25to35 = np.append(age_25to35, 0)
                age_35to45 = np.append(age_35to45, 0)
        
        return age_18to25, age_25to35, age_35to45, age_over45

    arrAges = generate_ages()
    #------------------------------------------------------------------

    
    #------------------------------------------------------------------
    #Generating Hours spent studying on campus...
    #------------------------------------------------------------------
    def gen_Minutes():
        min_1 = np.array([])
        min_2 = np.array([])
        min_3 = np.array([])

        for i in range(150):
            rand = 0
            rand = random.randint(1, 3)
            
            if (rand == 1):
                tMin = random.randint(60, 120)
                min_1 = np.append(min_1, tMin)
                min_2 = np.append(min_2, 0)
                min_3 = np.append(min_3, 0)
                
            elif(rand == 2):
                tMin = random.randint(120, 180)
                min_1 = np.append(min_1, 0)
                min_2 = np.append(min_2, tMin)
                min_3 = np.append(min_3, 0)
                
            elif(rand == 3):
                tMin = random.randint(240, 300)
                min_1 = np.append(min_1, 0)
                min_2 = np.append(min_2, 0)
                min_3 = np.append(min_3, tMin)
             
        return min_1, min_2, min_3        

    def covert_to_hours(arrMin):
        arrHour = np.array([])
        for x in arrMin:
            inHours = x / 60
            inHours = round(inHours, 2)
            arrHour = np.append(arrHour, inHours)
        return arrHour

    min_1to2, min_2to3, min_4to5 = gen_Minutes()

    hour_1to2 = covert_to_hours(min_1to2)
    hour_2to3 = covert_to_hours(min_2to3)
    hour_4to5 = covert_to_hours(min_4to5)
    #------------------------------------------------------------------


    #------------------------------------------------------------------
    #Generating Student mark achieved...
    #------------------------------------------------------------------
    def gen_percentage():
        marks = np.random.randint(0, 131, (150)) #0 - 130 marks

        lstperc = []
        for m in marks:
            perc = (m / 130) * 100
            lstperc.append(round(perc, 2))   
        return lstperc
    percentage = gen_percentage()
    #------------------------------------------------------------------


    #------------------------------------------------------------------
    #Generating Time taken on examination...
    #------------------------------------------------------------------
    time_Taken = np.random.randint(0, 180, (150)) #1 min - 180 min.

    #------------------------------------------------------------------


    #------------------------------------------------------------------
    #Populating dictionary for DataFrame Dataset...
    #------------------------------------------------------------------
    dictStudents = {"Student number": studNum, "Age 18-25": arrAges[0],
    "Age 25-35": arrAges[1], "Age 35-45": arrAges[2], "Over 45 age":arrAges[3],
    "1-2 Hours": hour_1to2, "2-3 Hours": hour_2to3, "4-5 Hours": hour_4to5,
    "Marks achieved": percentage, "Time taken": time_Taken}
    #------------------------------------------------------------------

    return dictStudents
#==============================================================================

dictExamn1 = create_DataFrame() #generate and call first dataset.
dictExamn2 = create_DataFrame() #generate and call second dataset.

#------------------------------------------------------------------
#Creating Dataset of the dictionary...
#------------------------------------------------------------------
data_frame1 = pd.DataFrame(dictExamn1)
data_frame2 = pd.DataFrame(dictExamn2)
#------------------------------------------------------------------

print("")
print("*********************************")
print("Examination One:")


```
