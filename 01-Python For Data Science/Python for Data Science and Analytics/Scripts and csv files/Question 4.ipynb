```python
#==================================================
#******************Question 4.1********************
#--------------------------------------------------
#Student Name: Sopumelela
#Student Surname: Sandekela
#Student Number: CON-1475940-C5L6
#==================================================

import numpy as np
import pandas as pd


#a.
#Read csv file.
dataframe = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

#b.
#Change them to lists.
BDW = dataframe.values.tolist()

#c.
#Populating BDW with first five rows.
BDW1 = []
icount = 0

print("Displaying the first five rows of BDW:")
print('-------------------------------------------------------------------------------')
for x in BDW:
    if icount >= 5:
        #Populating BDW1 with the other data.
        BDW1.append(x) #d.
    else:
        #Display five rows.
        line = ''
        for j in range(len(x)):
            line = line + str(round(x[j], 2)) +"  "
        print(line)
        print()
        
        icount += 1
print('-------------------------------------------------------------------------------')

print()
print()

#e.
print("Displaying first, second and third row of BDW1:")
print('-------------------------------------------------------------------------------')
for lst in BDW1[0:3]: #Slicing
    #Display first, second and third row of BDW1.
    line = ''
    for j in range(len(lst)):
        line = line + str(round(lst[j], 2)) +"  "
    print(line)
    print()
print('-------------------------------------------------------------------------------')
```

```python
#==================================================
#******************Question 4.2********************
#==================================================
import numpy as np
import pandas as pd

#Years = Keys | Values = Occurrence
indexcount_year = {}
#Years = Keys | Values = Indices
indexpercent_year = {}

years = []
for x in BDW1:
    years.append(round(x[1]))
    
#a.    
df1 = pd.Series(years).value_counts().reset_index()
df1.columns = ['Years', 'Frequency']

#Append the indexcount_year dictionary...
for y, f in zip(df1['Years'], df1['Frequency']):
    indexcount_year[''+str(y)+''] = f
    
#b.
#Append the indexpercent_year dictionary...
for lst in years:
    indexpercent_year[''+str(lst)+''] = years.index(lst)

#Prepare for output...
dfOutput1 = pd.Series(indexcount_year).reset_index()
dfOutput1.columns = ['Years', 'Frequency']

dfOutput2 = pd.Series(indexpercent_year).reset_index()
dfOutput2.columns = ['Years', 'Indices']


print("Display indexcount_year's content:")
dfOutput1
```

```python
#==================================================
#******************Question 4.3********************
#==================================================

def get_Columns(index):
    lst = []
    for i in range(len(BDW1)):
        lst.append(BDW1[i][index])
    return lst


#Getting index for the titles.
titles = []
for j in dataframe:
    titles.append(j)

#a
#.
index = titles.index('Math and Statistics')
Maths_Stats = get_Columns(index)
    
#.
index = titles.index('Physical Sciences')
Physic_Sci = get_Columns(index)

#.
index = titles.index('Engineering')
Engine = get_Columns(index)

#.
index = titles.index('Computer Science')
Comp_Sci = get_Columns(index)


print("Math and Statistics contents:")
print(Maths_Stats)
print()

print("Physical Sciences contents:")
print(Physic_Sci)
print()

print("Engineering contents:")
print(Engine)
print()

print("Computer Science contents:")
print(Comp_Sci)

print()
print()
print()

#b.
#Populating Year with every year.
index = titles.index('Year')
lst_year = get_Columns(index)

Year = []
for i in lst_year:
    Year.append(round(i))

print('All the years:')
print(Year)
```

```python
#==================================================
#******************Question 4.4********************
#==================================================
import numpy as np

#Convert to float...
def convert_to_float(lst):
    lstFloat = []
    for i in lst:
        lstFloat.append(i/1.0)
    return lstFloat

#Send to function to convert to float.
Maths_Stats = convert_to_float(Maths_Stats)
Physic_Sci = convert_to_float(Physic_Sci)
Engine = convert_to_float(Engine)
Comp_Sci = convert_to_float(Comp_Sci)

#a.
#Create numpy array Selected4Majors of the majors.
Selected4Majors = np.array([Maths_Stats, Physic_Sci, Engine, Comp_Sci])


#b.
#Creating the dictionary.
Majors = {'Math and Statistics': Selected4Majors[0], 
          'Physical Sciences': Selected4Majors[1], 
          'Engineering': Selected4Majors[2], 
          'Computer Science': Selected4Majors[3]}
#Output...
df = pd.DataFrame(Majors)
df
```

```python
#==================================================
#******************Question 4.5********************
#==================================================
import matplotlib.pyplot as plt
import numpy as np

#-------------------------------------------------------------------------
#a.
#Using for loop to plot each major.
#-------------------------------------------------------------------------
icount = 0
for major in Selected4Majors:
    if icount == 0:
        plt.plot(Year, major, color="blue", #Each major different color.
                 label='Math_Stats') #lebel for the legend
        icount += 1
        
    elif icount == 1:
        plt.plot(Year, major, color="red", 
                 label='Physic_Sci')
        icount += 1
        
    elif icount == 2:
        plt.plot(Year, major, color="green", 
                 label='Engine')
        icount += 1
        
    elif icount == 3:
        plt.plot(Year, major, color="purple", 
                 label='Comp_Sci')
        icount += 1
#-------------------------------------------------------------------------
#-------------------------------------------------------------------------
#c.
#Creating lables for the Graph.
#-------------------------------------------------------------------------
plt.xlabel("Year")
plt.ylabel("Selected4Degrees")
plt.title("Percentage of Selected4Degrees")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
#-------------------------------------------------------------------------
```
